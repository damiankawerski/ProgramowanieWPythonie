```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
```

## Mendell


```python
x = np.linspace(-2, 1, 600)
y = np.linspace(-1, 1, 400)

X, Y = np.meshgrid(x, y)

Z = X + Y * 1j  # tablica dwuwymiarowa wszystkich liczb zespolonych

s = Z.shape
mandelbrot = np.ones(s)
p = Z.copy()

start = time()

for i in range(255):
  mandelbrot[np.abs(Z) > 2] = i/2
  Z = Z ** 2 + p

end = time()

print(end - start)

plt.figure(figsize = (12, 8))
plt.imshow(mandelbrot, cmap = "jet")
```

```python
x = np.linspace(-2, 1, 1200)
y = np.linspace(-1, 1, 800)

X, Y = np.meshgrid(x, y)

Z = X + Y * 1j  # tablica dwuwymiarowa wszystkich liczb zespolonych

c = -0.73 + 0.19j

s = Z.shape
julia = np.ones(s)
p = Z.copy()


start = time()

for i in range(255):
  julia[np.abs(Z) > 2] = i/2
  Z = Z ** 2 + c

end = time()

print(end - start)

plt.figure(figsize = (12, 8))
plt.imshow(julia, cmap = "jet")
```
