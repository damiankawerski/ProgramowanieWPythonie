```python
from requests import get
from time import time

def stolica_dep(s):
    f = get("https://pl.wikipedia.org/wiki/" + s)  
    s = f.text
    f.close()
    return s.split('Stolica</a>', 1)[1].split('</a>', 1)[0].split('">', 1)[1].strip()

panstwa = ['Polska', 'USA', 'Czad', 'Hiszpania', 'Dania', 'Szwecja', 'Albania', 'Zimbabwe', 'Etiopia', 'Wietnam']

start = time()
for p in panstwa:
    print(f"{p}: {stolica_dep(p)}") 
end = time()
print(f"Czas wykonania: {end - start:.2f} sekundy")
```

```python
from requests import get
from time import time


def stolica(s):
    f = get("https://pl.wikipedia.org/wiki/" + s, stream=True)
    lines = f.iter_lines()

    for line in lines:
        line = line.decode("UTF-8")
        if "Stolica</a>" not in line:
            continue

        next(lines, None)
        next(lines, None)
       
        next_line = next(lines, "").decode("UTF-8")
        
        print(next_line.split('</a>', 1)[0].split('">', 1)[1].strip())
        return None
    

panstwa = ['Polska', 'USA', 'Czad', 'Hiszpania', 'Dania', 'Szwecja', 'Albania', 'Zimbabwe', 'Etiopia', 'Wietnam']

start = time()
for p in panstwa:
    print(f"{p}: {stolica(p)}") 
end = time()
print(f"Czas wykonania: {end - start:.2f} sekundy")
```

```python
from threading import Thread

watki = [ Thread(target=stolica, args=(p,)) for p in panstwa ]

start = time()
for w in watki:
    w.start()

for w in watki:
    w.join()
end = time()
print(end - start)

```
